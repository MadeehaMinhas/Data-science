```python
import pandas as pd
df=pd.read_csv('Combined_IoT_profiling_dataset_2022.csv')
df=df.iloc[:,1:]
df.head()
```

```python
devices=df['label'].unique()
len(devices)
```

```python
df[df['label']=='Audio_nestmini'].shape
```

```python
ratio=0.80
```

```python

i=0
for x in devices:
  df2=df[df['label']==x]
  print(x,df2.shape)
  df2 = df2.sample(frac=1)
  print(x,df2.shape)


  total_rows = df2.shape[0]
  #print(total_rows)
  train_size = int(total_rows*ratio)
  #print(train_size)
  # Split data into test and train

  train = df2[0:train_size]
  test = df2[train_size:]
  if i==0:
    train_df=pd.DataFrame(train,columns=df.columns)
    test_df=pd.DataFrame(test,columns=df.columns)
  else:
    train_df=pd.concat([train_df,train])
    test_df=pd.concat([test_df,test])
  print(x+'_train',train_df.shape)
  print(x+'_test',test_df.shape)
  i=i+1
```
